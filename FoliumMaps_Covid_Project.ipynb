**Folium** is built on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js (JavaScript) library. Simply, manipulate your data in Python, then visualize it on a leaflet map via Folium. Folium makes it easy to visualize data thatâ€™s been manipulated in Python, on an interactive Leaflet map.

Python provides folium as one of the powerful library

```
import folium
import pandas as pd
import numpy as np
df_usa=pd.DataFrame(pd.read_csv('usa_county_wise.csv'))
df_usa.head()
```

```
# get first 150 records
limit=150
df_usa=df_usa.iloc[0:limit,:]
df_usa.head()
```

```
print('plot any single data')
locationx=df_usa.head(1).Lat[0]
locationy=df_usa.head(1).Long_[0]
m=folium.Map(location=[locationx,locationy],tiles='Stamen Terrain',zoom_start=15)
m
```

```
print(' Add all locations ')
incidents=folium.map.FeatureGroup()

#add all maps
for lat,long in zip(df_usa.Lat,df_usa.Long_):
    
    incidents.add_child(folium.CircleMarker(location=[lat,long],radius=5,color='red',fill='True',fill_color='blue',transparency=0.6))

# add it to map
m.add_child(incidents)
m
```

```
print(' Add all locations with markers')
incidents=folium.map.FeatureGroup()

#add all maps
for long,lat,confirmed in zip(df_usa.Long_,df_usa.Lat,df_usa.Confirmed):
    covid=folium.Marker(location=[lat,long],popup=str(confirmed))
    
    incidents.add_child(covid)
# add it to map
m.add_child(incidents)
m
```
