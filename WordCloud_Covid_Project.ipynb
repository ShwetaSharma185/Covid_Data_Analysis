**Word Cloud** is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. Significant textual data points can be highlighted using a word cloud. Word clouds are widely used for analyzing data from social network websites.

```
# Lets try to show the COVID-19 symptoms using this great feature
import numpy as np
import pandas as pd
covid_symptoms=pd.read_csv('COVID-Symptoms-Raw-Data.csv')
covid_symptoms.head()
```

```
# Fetching the symptoms data from table
symptoms=covid_symptoms.Symptoms.values
experienced_symptoms=covid_symptoms.Experiencing_Symptoms.fillna("").values
print('We have got the symptoms from ',len(symptoms), ' different countries across globe')
```

Python too supports **file handling** and allows users to handle files i.e., to read and write files, along with many other file handling options, to operate on files. This concept here is also easy and short. Python treats file differently as text or binary and this is important. Each line of code includes a sequence of characters and they form text file. Each line of a file is terminated with a special character, called the EOL or End of Line characters like comma {,} or newline character. It ends the current line and tells the interpreter a new one has begun. Letâ€™s start with Reading and Writing files.


 There are three kinds of mode, that Python provides and how files can be opened:

**r**  for reading.

**w** for writing.

**a** for appending.

**r+** for both reading and writing

```
#Lets try to save all symptoms in one file and then read symptoms from that file and use for word cloud
file=open('symptoms_data.txt','w+')
symptoms_text=""

for sym in symptoms:
    file.write(sym+" ")

for sym in experienced_symptoms:
  file.write(sym+" ")

file.close()
```

```
# We are checking all the symptoms available from text file
sym_file=open('symptoms_data.txt','r')
symptoms_text=sym_file.read()*30
symptoms_text
```

```
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

World cloud is a **package/library** available in python. We will be using some key features like WordCloud, Stopwords and ImageColorGenerator

```
#Create a word cloud image
covid_wordcloud=WordCloud().generate(symptoms_text)
```

```
# Lets change some look and feel of image
covid_wordcloud=WordCloud(max_font_size=50,max_words=50,background_color='white').generate(symptoms_text)
plt.imshow(covid_wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

**Word cloud** in python also provides the feature of masking image. So various social sites and ventures widely use this feature for advertising the key factors of their business to attract more consumers.

Python provides lots of libraries for image processing, including âˆ’ OpenCV âˆ’ Image processing library mainly focused on real-time computer vision with application in wide-range of areas like 2D and 3D feature toolkits, facial & gesture recognition, Human-computer interaction, Mobile robotics, Object identification and others.

```
mask=np.array(Image.open('redheart.jpg'))
```

```
plt.figure(figsize=(7,7))
wordcloud_mask=WordCloud(background_color='White',mode='RGBA',mask=mask).generate(symptoms_text)
image_colors=ImageColorGenerator(mask)

plt.imshow(wordcloud_mask.recolor(color_func=image_colors),interpolation='bilinear')
plt.axis('off')
plt.show()
```
