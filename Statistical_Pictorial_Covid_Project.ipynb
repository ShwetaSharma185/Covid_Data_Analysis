**Matplotlib** is the brainchild of **John Hunter** (1968-2012), who, along with its many contributors, have put an immeasurable amount of time and effort into producing a piece of software utilized by thousands of scientists worldwide.

Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.

One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
patient_data=pd.read_csv('patients_data.csv')
patient_data.head()
```

```
print('TOP 10 STATE MOST AFFECTED BY COVID-19')
plt.figure(figsize=(15,10))
patient_data.groupby('Detected State').size().nlargest(10).plot.bar()
plt.xticks(rotation=50)
plt.xlabel('Affected State')
plt.ylabel('No of patients')
plt.show()
```

```

plt.style.use('ggplot')
patient_data['Age Bracket'].value_counts().head().plot.bar()
plt.title('TOP 5 age group MOST affected')
plt.xlabel('Age')
plt.ylabel('No of patients affected')
plt.show()
```

```
plt.title('TOP 10 City having maximum COVID-19 cases')
patient_data.groupby('Detected State')['Detected City'].value_counts().nlargest(10).plot.bar()
plt.xticks(rotation=90)
plt.show()
```

```
patient_data.groupby('Detected State')['Detected City'].value_counts().nlargest(10)
```

**Seaborn** is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Seaborn is an open source, BSD-licensed Python library providing high level API for visualizing the data using Python programming language.



```
import seaborn as sns
```

## **Pictorial View of Patients Status**


```
sns.catplot(x="Current Status",y='Patient Number' ,kind="bar", data=patient_data,height=4,aspect=2)
```

**Gender Statistic of Patients**

```

sns.catplot(x="Gender",y='Patient Number' , palette="ch:.25", data=patient_data);
```

State-Wise Statistics

```
f, ax = plt.subplots(figsize=(10, 12))
sns.countplot(y="Detected State", data=patient_data);
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
country_data=pd.read_csv('country_wise_latest.csv')
country_data.head()
```

**COUNTRY_WISE CONFIRMED CASES**

```
x=country_data.sort_values(by='Confirmed',ascending=False)['Country/Region'].head(10)
y=country_data.sort_values(by='Confirmed',ascending=False)['Confirmed'].head(10)
plt.figure(figsize=(12,7))
plt.plot(x,y)
plt.xlabel('Country')
plt.ylabel('Patients in Lakhs')
plt.show()
```

**TOP 20 CITY STATISTICS USING SEABORN**

```
f, ax = plt.subplots(figsize=(10, 12))
sns.countplot(y="Detected City", order=patient_data['Detected City'].value_counts().nlargest(20).index, data=patient_data);
```

STATISTICAL PLOT OF HOW THE DEATHS AND CONFIRMED CASES ARE BEEN LINKED. 

ALSO MORE DARKER HEXAGON MORE CLOSELY ARE RELATED

```
country_data=country_data.sort_values(by='Confirmed',ascending=False)
sns.jointplot(country_data['Confirmed'][:30],
              country_data['Deaths'][:30],
              kind='hex',data=country_data)
plt.show()
```

**NEW CASES vs NEW DEATHS Point Plot**

```
country_data=country_data.sort_values(by='Confirmed',ascending=False)
sns.pointplot(country_data['New cases'].value_counts().nlargest(50),
              country_data['New deaths'].value_counts().nlargest(50),
              data=country_data)
plt.show()
```

## **TOP 5 COUNTRIES MOST AFFECTED WITH COVID**

```
country_data=country_data[['Country/Region','Confirmed']].sort_values(by='Confirmed',ascending=False).head()
plt.figure(figsize=(8,8))
plt.pie(country_data.Confirmed,labels=country_data['Country/Region'])
plt.show()
```
